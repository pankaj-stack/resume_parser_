```python
from tkinter import *
from tkinter import filedialog
from tkinter import messagebox
import os
import shutil
import docxpy
import time
from PIL import Image, ImageTk
import re

class resume:
    def __init__(self):
        self.win=Tk()
        self.counter=0
        
        self.win.geometry(f'{self.win.winfo_screenwidth()}x{self.win.winfo_screenheight()}')
        self.win.config(bg='#856ff8')
        
        self.directory_path=Label(self.win,text='Directory Path',bg='green',
                                 fg='white')
        self.directory_path.grid(row=0,column=0,padx=10,pady=5)
        
        self.folder_path=StringVar()
        self.directory_entry=Entry(self.win,textvariable=self.folder_path)
        self.directory_entry.grid(row=0,column=1,pady=5,padx=10)

        self.enter_keyword=Label(self.win,text='Enter Keyword',bg='green',
                                fg='white')
        self.enter_keyword.grid(row=1,column=0,pady=5,padx=10)
        
        self.parser=StringVar()
        self.parser_entry=Entry(self.win,textvariable=self.parser)
        self.parser_entry.grid(row=1,column=1,pady=5,padx=10)

        self.browse=Button(self.win,text='Browse'
                           ,bg='blue',fg='white',command=self.file_opener)
        self.browse.grid(row=0,column=2,pady=5,padx=10)
        
        self.search=Button(self.win,text='Search'
                           ,bg='blue',fg='white',command=self.search_keyword)
        self.search.grid(row=1,column=2,pady=5,padx=10)
        
        self.exit_button=Button(self.win,text='Exit',width=10,bg='red',
                            fg='white',command=self.exit_window)
        self.exit_button.grid(row=2,column=0,pady=5,padx=10)
        
        #Reset Button
        self.reset_button=Button(self.win,text='Reset',bg='blue',fg='white',command=self.reset)
        self.reset_button.grid(row=2,column=1,pady=5,padx=10)
        
        self.open_button=Button(self.win,text='    Open    ',bg='blue',fg='white',command=self.listbox_file_opener)
        self.open_button.grid(row=2,column=3,pady=5,padx=10)
        self.label_image()
        
       
        self.win.mainloop()
    def label_image(self):
        self.img=Image.open('resume_parser_edited.png')
        self.img=self.img.resize((int(self.win.winfo_screenwidth()/2),int(self.win.winfo_screenheight()/2)))
        self.img=ImageTk.PhotoImage(self.img)
        self.img_label=Label(image=self.img,bg='white')
        self.img_label.image=self.img
        self.img_label.grid(row=3,column=0,columnspan=2)
 
        self.img=Image.open('resume_parser_2.png')
        self.img=self.img.resize((int(self.win.winfo_screenwidth()/2),int(self.win.winfo_screenheight()/2)))
        self.img=ImageTk.PhotoImage(self.img)
        self.img_label=Label(image=self.img,bg='white')
        self.img_label.image=self.img
        self.img_label.grid(row=3,column=2,columnspan=2)
        
        

    def set_entry_text(self,text):
        self.text=text
        self.directory_entry.delete(0,END)
        self.directory_entry.insert(0,text)
        return

    def file_opener(self):
        # Allow user to select a directory and store it in global var
        # called folder_path
        self.filename = filedialog.askdirectory()
        self.folder_path.set(self.filename)
        print(self.filename)
        self.set_entry_text(self.filename)
        self.to_color_window()
        return self.filename

    def exit_window(self):
        self.win['bg']='red'
        self.win.destroy()
    def to_color_window(self):
            self.win['bg']='#800000'
            self.search['bg']='blue'
            self.search['fg']='white'
            self.browse['bg']='blue'
            self.browse['fg']='white'
            self.directory_path['bg']='green'
            self.directory_path['fg']='white' 
            self.enter_keyword['bg']='green'
            self.enter_keyword['fg']='white'
            self.open_button['bg']='blue'
            self.open_button['fg']='white'
    def search_keyword(self):
        self.doc_docx_file=[]
        self.matched_files=[]
        try:
            self.file_list=os.listdir(self.filename)
        except AttributeError as e:
            messagebox.showerror('Error','First Select the path')
        except FileNotFoundError as e:
            messagebox.showerror('Error','Pleae specify the valid path')
        for i in self.file_list:
            if i.endswith('doc') or i.endswith('docx'):
                self.doc_docx_file.append(i)
                self.text=docxpy.process(self.filename+'/'+i)
                
                if len(self.parser.get())==1:
                    if self.parser.get()=='a':
                        self.searched_item=[]
                        continue
                    else:
                        pattern=re.compile(f'\s+{self.parser.get().lower()}\s+')
                else:
                    pattern=re.compile(f'\s{self.parser.get().lower()}\s?')

                self.searched_item=re.findall(pattern,self.text.lower())
                try:
                    if self.searched_item[0].lower() in self.text.lower():
                        self.matched_files.append(i)
#                         print(f'matched files is : {self.matched_files}')
                except IndexError as e:
                    pass
        if len(self.matched_files)!=0:
            if os.path.exists('C:/Users/Dell/Desktop'+'/'+self.parser.get()):
                        if self.parser.get()!='':
                            print(self.searched_item)
                            messagebox.showerror('Error','This folder is already exist')
  
        #             sys.exit()
            else:
                os.mkdir('C:/Users/Dell/Desktop'+'/'+self.parser.get())
                for i in self.matched_files:
                    shutil.copy(self.filename+'/'+i,'C:/Users/Dell/Desktop'+'/'+self.parser.get())
                messagebox.showinfo('INFO',f"Your file is moved into the {'C:/Users/Dell/Desktop/'}{self.parser.get()}")
        if self.parser.get()!='':
            self.list_box=Listbox(self.win,width=35,selectmode=EXTENDED,
                              bg='green',fg='white')
        
            self.list_box.insert(0,f'Toatal files found : {len(self.file_list)}')
            self.list_box.insert(1,f'Total Docx files found : {len(self.doc_docx_file)}')
            self.list_box.insert(2,f'Total Matched files : {len(self.matched_files)}')
            for j in range(len(self.matched_files)):
                self.list_box.insert(j+3,f'{self.matched_files[j]}')
            self.list_box.grid(row=1,column=3,rowspan=1)
            self.win['bg']='yellow'
        
            
    def listbox_file_opener(self):
        file=self.list_box.curselection()
#         print(f' value of file is {file}')
        for item in file:
            try:
                os.startfile('C:/Users/Dell/Desktop'+'/'+self.parser.get()+'/'+self.list_box.get(item))
            except FileNotFoundError as e:
                messagebox.showerror('Error','Please select the suitable file')
        
    def reset(self):
        try:
            self.list_box.delete(0,END)
        #         self.folder_path.set('')
            self.folder_path.set('')
            self.parser.set('')
            self.list_box['bg']='white'
            self.win['bg']='black'
            self.filename=''
            self.search['bg']='black'
            self.search['fg']='white'
            self.browse['bg']='black'
            self.browse['fg']='white'
            self.directory_path['bg']='black'
            self.directory_path['fg']='white' 
            self.enter_keyword['bg']='black'
            self.enter_keyword['fg']='white'
            self.open_button['bg']='black'
            self.open_button['fg']='white'
        except AttributeError as e:
                pass

obj=resume()
```
